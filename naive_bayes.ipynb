```python
class NaiveBayes:
    def __init__(self, features_df_train, target_df_train, df_train):
        self.df_train = df_train
        self.features_df_train = features_df_train
        self.target_df_train = target_df_train
        self.nfeatures = len(features_df_train.columns)
        self.msamples = len(features_df_train)
        self.feature_names = features_df_train.columns.to_list()
        self.target_name = list(target_df_train.name)
        self.prob_f_to_fc_to_tc = {}
        self.prob_f_to_fc = {}


        self.prior()		
        self.evidence()
        self.likelihood()
        # self.posterior()
    
    def prior(self):
        self.prob_TC_ = self.target_df_train.value_counts(normalize = True).to_dict()
    
    def likelihood(self):
        for feature_name in self.feature_names:
            temp_df = self.df_train.groupby([target_name[0], feature_name])[feature_name].count().unstack()
            self.prob_f_to_fc_to_tc[feature_name] = temp_df.div(temp_df.sum(axis= 1), axis =0).to_dict()

    def evidence(self):
        for feature_name in self.feature_names:
            temp_df_2 = self.df_train.groupby(feature_name)[feature_name].count()
            self.prob_f_to_fc[feature_name] = temp_df_2.div(temp_df_2.sum()).to_dict()

    def posterior(self, ficj):
        # ficj = list(ficj)
        self.likelihood_p = {}
        self.evidence_p = {}
        self.post = {}
        self.prior_p = {}
        i=0
        
        for tck in self.prob_TC_.keys():


            # Likelihood
            self.likelihood_p[tck] = 1
            for fi in self.feature_names:
                ficj_i = ficj[i]
                self.likelihood_p[tck] *= self.prob_f_to_fc_to_tc[fi][ficj_i][tck]
                i +=1

            # Prior
            self.prior_p[tck] = self.prob_TC_[tck]
            i = 0

            # Evidence
            self.evidence_p[tck] = 1
            for fi in self.feature_names:
                ficj_i = ficj[i]
                self.evidence_p[tck] *= self.prob_f_to_fc[fi][ficj_i]
                i+=1
            
            i=0
            
            self.post[tck] = self.likelihood_p[tck] * self.prior_p[tck] / self.evidence_p[tck]

    
    def predict(self):
        probable_class = max(self.post, key = self.post.get)
        return probable_class



        #P(TC1| f1c1, f2c3, f3c4, f4c1) 
        # = prob_f_to_fc_to_tc[f1][f1c1][tc1] * prob_f_to_fc_to_tc[f2][f2c3][tc1] 
        # * prob_f_to_fc_to_tc[f3][f3c4][tc1] * prob_f_to_fc_to_tc[f4][f4c1][tc1]
        # * prob_TC_[tc1]

    def accuracy_score(self):
        ncorrect_pred = 0

        for i in range(0, len(self.features_df_train), 1):
            self.posterior(self.features_df_train.iloc[i])
            prob_class = self.predict()
            if prob_class == self.target_df_train[i]:
                ncorrect_pred +=1
        
        self.accuracy = ncorrect_pred / len(self.target_df_train)
        

    
        return self.accuracy
        
    
```
